```
#romberg integration

from numpy import *
from math import *

def f(x):
  return cos(x)**2

def trapezoid(a, b, n):
  h=(b-a)/n
  s=0
  for i in range(n-1):
    s += f(a+h*(i+1))
  return h/2*(f(a)+2*s+f(b))

trap = zeros((order,1))
for i in range(order):
    trap[i,0] = trapezoid(0,pi/4,2**i)
print("Table Trapezoid")
print(trap)
print("Result integration of cos(x)**2 using trapezoid rule: ", "%f" % trap[order-1][0])
```

```
order = 4
#create matrix of zero with size order x order
rom = zeros((order, order))
#get trapezoid rule for first column
def romberg(a,b):
  for i in range(order):
    rom[i, 0] = trapezoid(a,b,2**i)

  for i in arange(1, order):
    for j in arange(i, order):
      rom[j,i] = rom[j,i-1]+(rom[j,i-1]-rom[j-1,i-1])/(4**i-1)
  return rom
print("Table Romberg")
print(romberg(0, pi/4))
print("Result integration of cos(x)**2 using Romberg Integration: ", "%f" % rom[order-1][order-1])
```
